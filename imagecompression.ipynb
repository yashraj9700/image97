```python
import cv2
import numpy as np
import os
from matplotlib import pyplot as plt

# Load the original image
image_path = '/content/dog.jpeg'  # Update this to your actual image file path
image = cv2.imread(image_path)

# Save the image with lossless and lossy compression
cv2.imwrite('image_lossless.png', image)  # Lossless PNG
cv2.imwrite('image_lossy.jpg', image, [cv2.IMWRITE_JPEG_QUALITY, 50])  # Lossy JPEG with 50% quality

# Load images for comparison
image_original = cv2.imread(image_path)
image_lossless = cv2.imread('image_lossless.png')
image_lossy = cv2.imread('image_lossy.jpg')

# Convert images from BGR to RGB
image_original_rgb = cv2.cvtColor(image_original, cv2.COLOR_BGR2RGB)
image_lossless_rgb = cv2.cvtColor(image_lossless, cv2.COLOR_BGR2RGB)
image_lossy_rgb = cv2.cvtColor(image_lossy, cv2.COLOR_BGR2RGB)

# Plot the images
plt.figure(figsize=(15, 10))

plt.subplot(1, 3, 1)
plt.title('Original Image')
plt.imshow(image_original_rgb)
plt.axis('off')

plt.subplot(1, 3, 2)
plt.title('Lossless Compression (PNG)')
plt.imshow(image_lossless_rgb)
plt.axis('off')

plt.subplot(1, 3, 3)
plt.title('Lossy Compression (JPEG)')
plt.imshow(image_lossy_rgb)
plt.axis('off')

plt.show()

# Compare Dimensions
height_original, width_original, _ = image_original.shape
height_lossless, width_lossless, _ = image_lossless.shape
height_lossy, width_lossy, _ = image_lossy.shape

print(f"Dimensions of Original Image: Width = {width_original}, Height = {height_original}")
print(f"Dimensions of Lossless Compression (PNG) Image: Width = {width_lossless}, Height = {height_lossless}")
print(f"Dimensions of Lossy Compression (JPEG) Image: Width = {width_lossy}, Height = {height_lossy}")

# Compare dimensions
if (width_original, height_original) == (width_lossless, height_lossless) == (width_lossy, height_lossy):
    print("All images have the same dimensions.")
else:
  print("The images have different dimensions.")

# Calculate Memory Usage
def calculate_memory_usage(image):
    height, width, channels = image.shape
    return height * width * channels  # Each channel is 1 byte for standard images

memory_original = calculate_memory_usage(image_original)
memory_lossless = calculate_memory_usage(image_lossless)
memory_lossy = calculate_memory_usage(image_lossy)

print(f"Memory usage of Original Image: {memory_original / (1024 * 1024):.2f} MB")
print(f"Memory usage of Lossless Compression (PNG) Image: {memory_lossless / (1024 * 1024):.2f} MB")
print(f"Memory usage of Lossy Compression (JPEG) Image: {memory_lossy / (1024 * 1024):.2f} MB")

# Compare File Sizes
import os

# File paths
file_original = image_path
file_lossless = 'image_lossless.png'
file_lossy = 'image_lossy.jpg'

# Get file sizes in bytes
size_original = os.path.getsize(file_original)
size_lossless = os.path.getsize(file_lossless)
size_lossy = os.path.getsize(file_lossy)

print(f"File size of Original Image: {size_original / 1024:.2f} KB")
print(f"File size of Lossless Compression (PNG) Image: {size_lossless / 1024:.2f} KB")
print(f"File size of Lossy Compression (JPEG) Image: {size_lossy / 1024:.2f} KB")
```
