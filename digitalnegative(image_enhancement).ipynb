```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

# Step 1: Create a binary matrix (0s and 255s)
binary_matrix = np.zeros((100, 100), dtype=np.uint8)  # Create a black matrix
binary_matrix[30:70, 30:70] = 255  # Create a white square

# Step 2: Display the original binary image
plt.figure(figsize=(10, 5))
plt.subplot(1, 3, 1)
plt.title("Original Binary Image")
plt.imshow(binary_matrix, cmap='gray')
plt.axis('off')

# Step 3: Adjust brightness and contrast
def adjust_brightness_contrast(image, brightness=0, contrast=1):
    adjusted = cv2.convertScaleAbs(image, alpha=contrast, beta=brightness)
    return adjusted

# Adjust settings
brightness = 50  # Increase brightness
contrast = 1.0   # No change in contrast
adjusted_image = adjust_brightness_contrast(binary_matrix, brightness, contrast)

# Step 4: Create a digital negative
negative_image = 255 - adjusted_image
# Step 5: Plot the results
plt.subplot(1, 3, 2)
plt.title("Adjusted Image")
plt.imshow(adjusted_image, cmap='gray')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.title("Digital Negative")
plt.imshow(negative_image, cmap='gray')
plt.axis('off')

plt.show()
```
